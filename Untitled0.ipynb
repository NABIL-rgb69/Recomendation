```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import ipywidgets as widgets
from IPython.display import display, clear_output
from google.colab import files
from sklearn.feature_extraction.text import CountVectorizer






uploaded = files.upload()


file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

df.head()
```

```python
df['genres_str'] = df['listed_in']
df['duration_int'] = df['duration'].str.extract(r'(\d+)').astype(float)
```

```python
status_output = widgets.Output()
tombol_cluster = widgets.Button(description="Generate Cluster")

def generate_cluster(b):
    with status_output:
        clear_output()
        print("‚è≥  Memproses cluster...")

        # Hapus kolom Cluster lama kalau ada
        if 'Cluster' in df.columns:
            df.drop(columns=['Cluster'], inplace=True)

        # vektorisasi genre + gabung dengan fitur numerik
        vectorizer = CountVectorizer(tokenizer=lambda x: x.split(', '))
        genre_matrix = vectorizer.fit_transform(df['genres_str'].fillna(""))
        numerik = df[['duration_int', 'release_year']].fillna(0)
        gabung = np.hstack((genre_matrix.toarray(), numerik.values))

        # scaling + clustering
        scaler = StandardScaler()
        X = scaler.fit_transform(gabung)
        kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
        df['Cluster'] = kmeans.fit_predict(X) + 1  # ‚Üí hasilnya 1‚Äì4

        clear_output()
        print("‚úÖ Cluster berhasil dibuat (1‚Äì4)!")

tombol_cluster.on_click(generate_cluster)
display(tombol_cluster, status_output)
```

```python
input_user = widgets.Text(placeholder='Masukkan nama user')
dropdown_film = widgets.Dropdown(
    options=sorted(df['title'].dropna().unique()),
    description='Pilih Film:'
)
tombol_rekom = widgets.Button(description="Tampilkan Rekomendasi")
output_rekom = widgets.Output()

def show_rekom(b):
    with output_rekom:
        clear_output()

        if 'Cluster' not in df.columns:
            print("‚ö†Ô∏è  Harap klik tombol 'Generate Cluster' terlebih dahulu.")
            return

        user = input_user.value
        film = dropdown_film.value

        # Cluster utk film yg dipilih
        c = df[df['title']==film]['Cluster'].values[0]
        genre_pilihan = df[df['title']==film]['genres_str'].values[0]

        print(f"üë§ User  : {user}")
        print(f"üé¨ Film yang dipilih : {film}")
        print(f"üìä Cluster : {c}  (artinya film ini berada pada Cluster 1‚Äì4)")
        print(f"üé≠ Genre  : {genre_pilihan}")
        print("\n‚úÖ Film-film berikut direkomendasikan karena berada pada cluster & genre yang sama:\n")

        rekom = df[(df['Cluster']==c) & (df['title'] != film)]
        display(rekom[['title','genres_str','release_year','duration_int']].head(5))

tombol_rekom.on_click(show_rekom)

display(input_user, dropdown_film, tombol_rekom, output_rekom)
```
